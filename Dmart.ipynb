**Key Observation:-**

*   The primary goal is to fill in missing values in the "Brand" column of the dataset.

*  Many products might have their brand name missing, and this code aims to extract a potential brand name from the product name itself.

*   Adding "No brand"  value in the rows for those product which don't have brand name.





```python
existing_brands = df["Brand"].dropna().unique().tolist()

def extract_brand(x):
  if pd.isnull(x["Brand"]):
    first_word=x["Name"].split()[0]
    if first_word in existing_brands:
      return first_word
    else:
      return "No brand"
  else:
    return x["Brand"]


df["Brand"]=df.apply(extract_brand,axis=1)
```

### **Key Observation:-**


*   Since each row seems to represent a distinct sale, the "Discounted Price" would likely reflect the final price paid for that specific product and quantity combination.

*   If the quantity for each unique sale is already incorporated into the "Discounted Price" (e.g., a bulk discount for a larger size), then multiplying by quantity again would be redundant and inaccurate.

*   In this specific case, copying "Discounted Price" to "Revenue" could be a reasonable simplification, as it aligns with the observation that each sale is unique.




```python
df['Revenue']=df["DiscountedPrice"].copy()
```

### **Key Observation:-**

*   The purpose of this code is to determine the discount percentage applied to each product based on its original price (Price) and its discounted price (DiscountedPrice).




```python
df["DiscountPercentage"]= round(((df["Price"]-df["DiscountedPrice"])/df["Price"])*100,0)
#df["DiscountPercentage"] = df["DiscountPercentage"].astype(str) + "%"
```

# **Visualization**

```python
Top_Brand_sales = df.groupby("Brand")["Revenue"].sum().drop("No brand").sort_values(ascending=False).head(20)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=Top_Brand_sales.values, y=Top_Brand_sales.index,palette='Blues_r')
plt.title('Top 20 Brands by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Brands')
plt.xticks(rotation=45)
plt.show()
```

```python
Bottom_Brand_sales=df.groupby("Brand")["Revenue"].sum().nsmallest(20).sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=Bottom_Brand_sales.values, y=Bottom_Brand_sales.index,palette='Blues_r')
plt.title('Bottom 20 Brands by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Brands')
plt.xticks(rotation=45)
plt.show()
```

```python
Category_sales=df.groupby("Category")["Revenue"].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=Category_sales.values, y=Category_sales.index,palette='magma')
plt.title('Categories by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Categories')
plt.xticks(rotation=45)
plt.show()
```
